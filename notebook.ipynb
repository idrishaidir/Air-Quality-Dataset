## Import Semua Packages/Library yang Digunakan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
changping_df = pd.read_csv("https://raw.githubusercontent.com/marceloreis/HTI/refs/heads/master/PRSA_Data_20130301-20170228/PRSA_Data_Changping_20130301-20170228.csv")
changping_df.head()
```

**Insight:**
- terdapat beberapa jenis polutan dan temperatur
- terdapat kondisi cuaca seperti rain.

### Assessing Data

```python
changping_df.info()
changping_df.isna().sum()
```

```python
print(f"Jumlah Duplikasi:{changping_df.duplicated().sum()}")
changping_df.describe()
```

**Insight:**
- terdapat banyak missing value
- year,month, day, hour bertipe data int

### Cleaning Data

```python
#Missing Value
changping_df.interpolate(method='linear', inplace=True)
changping_df.wd.fillna(changping_df.wd.mode()[0], inplace=True)

changping_df.isna().sum()
```

```python
changping_df['Tanggal'] = pd.to_datetime(changping_df[['year', 'month', 'day', 'hour']])
changping_df.drop(columns=['month', 'day', 'hour'], inplace=True)
changping_df.head()
```

```python
changping_df.max().head()
```

**Insight:**
- tidak ada missing value
- terjadi keanehan pada kadar polusi (PM2.5, PM10, NO2)

## Exploratory Data Analysis (EDA)

### Explore ...

```python
changping_df.describe(include="all")
```

```python
teliti = ["PM2.5", "PM10", "NO2", "TEMP", "RAIN"]

for col in teliti:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(changping_df[col], kde=True, bins=50, ax=axes[0])
    axes[0].set_title(f"Distribusi {col}")
    sns.boxplot(y=changping_df[col], ax=axes[1])
    axes[1].set_title(f"Boxplot {col}")
    plt.show()
```

```python
for col in teliti:
    plt.figure(figsize=(8, 4))
    changping_df.groupby(changping_df.year)[col].mean().plot(marker="o")
    plt.title(f"Tren Rata-rata {col} per Tahun")
    plt.xlabel("Tahun")
    plt.ylabel(col)
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(changping_df[teliti].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matriks Korelasi")
plt.show()

sns.pairplot(changping_df[teliti])
plt.show()
```

```python
filter = ["PM2.5", "PM10", "NO2"]
for col in filter:
    Q1 = changping_df[col].quantile(0.25)
    Q3 = changping_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"{col} -> Batas bawah: {lower_bound}, Batas atas: {upper_bound}")
    changping_df = changping_df[(changping_df[col] >= lower_bound) & (changping_df[col] <= upper_bound)]
```

```python
for col in teliti:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(changping_df[col], kde=True, bins=50, ax=axes[0])
    axes[0].set_title(f"Distribusi {col}")
    sns.boxplot(y=changping_df[col], ax=axes[1])
    axes[1].set_title(f"Boxplot {col}")
    plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:Apakah hujan (RAIN) berpengaruh terhadap penurunan kadar polutan di udara?

```python
rain_pollution = changping_df.groupby('RAIN')[['PM2.5', 'PM10', 'NO2']].mean().reset_index()

rain_pollution_melted = rain_pollution.melt(id_vars='RAIN', var_name='Polutan', value_name='Kadar')

plt.figure(figsize=(8, 6))
sns.barplot(x='Polutan', y='Kadar', hue='RAIN', data=rain_pollution_melted)
plt.xlabel("Jenis Polutan")
plt.ylabel("Kadar Rata-rata")
plt.title("Perbandingan Kadar Polutan antara Kondisi Hujan dan Tidak Hujan")
plt.show()
```

### Pertanyaan 2:Bagaimana hubungan antara suhu (TEMP) dan tingkat polutan (PM2.5, PM10, NO2)?

```python
corr_temp = changping_df[["TEMP", "PM2.5", "PM10", "NO2"]].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_temp, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korelasi antara Suhu dan Polutan")
plt.show()
```

**Insight:**
- ketika terjadi hujan kecendurangan polutan akan turun meskipun tidak signifikan
- temperatur memiliki hubungan yang sangat lemah dengan NO2 DAN PM2.5
